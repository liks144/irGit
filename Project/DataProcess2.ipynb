```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
%matplotlib inline
sns.set_style('whitegrid')
```

```python
def trans(x):
    if x <= 5:
        return x
    if 5 < x <= 100:
        return 100
    if 100 < x <= 200:
        return 200
    if 200 < x <= 300:
        return 300
    if 300 < x <= 400:
        return 400
    if 400 < x <= 500:
        return 500
    if 500 < x <= 600:
        return 600
    if x > 600:
        return 700
```

```python
import pickle

def save_obj(obj, name ):
    with open('obj/'+ name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)

def load_obj(name ):
    with open('obj/' + name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
answer_info = pd.read_csv('data_set_0926/answer_info_0926.txt', header=None, sep='\t')
answer_info.columns = ['回答ID','问题ID','用户ID','回答创建时间', '回答内容的单字编码序列','回答内容的切词编码序列', '回答是否被标优', '回答是否被推荐', '回答是否被收入圆桌', '是否包含图片', '是否包含视频', '回答字数', '点赞数','取赞数','评论数','收藏数', '感谢数','举报数','没有帮助数', '反对数']
print(answer_info.columns)

question_info = pd.read_csv('data_set_0926/question_info_0926.txt', header=None, sep='\t')
question_info.columns = ['问题ID','问题创建时间', '问题标题的单字编码序列',  '问题标题的切词编码序列',  '问题描述的单字编码序列',  '问题描述的词编码序列',  '问题绑定的话题ID']
print(question_info.columns)

data = pd.merge(answer_info, question_info, how='left', on='问题ID')
print(data.columns)
```

```python
member_info = pd.read_csv('data_set_0926/member_info_0926.txt', header=None, sep='\t')
member_info.columns = ['用户id','性别','创作关键词','创作数量等级','创作热度等级','注册类型','注册平台','访问评率','用户二分类特征a','用户二分类特征b','用户二分类特征c','用户二分类特征d','用户二分类特征e','用户多分类特征a','用户多分类特征b','用户多分类特征c','用户多分类特征d','用户多分类特征e','盐值','关注话题','感兴趣话题']
print(member_info.columns)
```

```python
del member_tf_extend
member_tf_extend = load_obj('member_topicfreq')
```

```python
member_interest = {}
for index, row in member_info.iterrows():
#将关注话题集合加入到回答问题涉及的话题集合中，规则为没有置1，有则加1
    if row['关注话题'] != '-1':
        if row['用户id'] not in member_tf_extend:
            member_tf_extend[row['用户id']] = dict(zip(row['关注话题'].split(','),[1]*len(row['关注话题'].split(','))))
        else:
            for topic in row['关注话题'].split(','):
                if topic in member_tf_extend[row['用户id']]:
                    member_tf_extend[row['用户id']][topic] += 1;
                else:
                    member_tf_extend[row['用户id']][topic] = 1;
    else:
        member_tf_extend[row['用户id']] = {}

#统计感兴趣的话题
    if row['用户id'] in member_interest:
        print('用户重复')
        
    member_interest[row['用户id']] = {}
    if row['感兴趣话题'] != '-1':
        item_list = row['感兴趣话题'].split(',')
        for i in item_list:
            a,b=i.split(':')
            member_interest[row['用户id']][a] = float(b)
        
save_obj(member_tf_extend,'member_tf_extend')
save_obj(member_interest,'member_interest')
```

```python
# 统计 用户-回答涉及话题
member_topic = {}
for index, row in data.iterrows():
    if row['用户ID'] in member_topic:
        member_topic[row['用户ID']].update(row['问题绑定的话题ID'].split(","))
    else :
        member_topic[row['用户ID']] = set(row['问题绑定的话题ID'].split(","))
```

```python
# 统计 用户-回答涉及话题及频率
member_topicfreq = {}
member_answerfreq = {}
for index, row in data.iterrows():
    if row['用户ID'] in member_topicfreq:
        for topic in row['问题绑定的话题ID'].split(","):
                member_topicfreq[row['用户ID']][topic] += 1;
                member_answerfreq[row['用户ID']] += 1;
    else :
        member_topicfreq[row['用户ID']] = dict(zip(member_topic[row['用户ID']],[1]*len(member_topic[row['用户ID']])))
        member_answerfreq[row['用户ID']] = 1;
```

```python
save_obj(member_topicfreq,'member_topicfreq')
```
